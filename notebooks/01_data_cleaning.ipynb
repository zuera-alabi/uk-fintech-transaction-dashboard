## Data Cleaning
In this section we address the issues identified during 
data understanding, fixing data types and handling 
missing values

#### convert transaction_ date to datetime

```python
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
print(df['transaction_date'].dtype)
```

```python
df.head()
```

#### Handling missing values 

```python
print("Missing values before cleaning:")
print(df.isnull().sum()[df.isnull().sum() > 0])
```

##### Replace transaction_category with Uncategorised 
**Reasoning:** The transaction_category column contained 
225 missing values (1.5% of the dataset). Since these 
transactions exist but lack a category label, they were 
filled with 'Uncategorised' rather than being dropped or 
filled with an assumed category. This approach preserves 
all 15,000 records while maintaining transparency about 
the data gap. The small percentage means this decision 
has minimal impact on the overall category analysis.

```python
df["transaction_category"] = df["transaction_category"].fillna("Uncategorised")
print(f"transaction_category missing: {df['transaction_category'].isnull().sum()}")
df
```

##### Fill payment method with mode
**Reasoning:** The payment_method column contained 150 
missing values (1.0% of the dataset). Since payment method 
is a categorical column with no numerical meaning, it was 
filled with the mode (most frequently occurring value). 
This is a standard approach for categorical missing values 
where the missingness appears random rather than systematic. 
The small percentage ensures minimal impact on payment 
method analysis.

```python
mode_payment = df['payment_method'].mode()[0]
df['payment_method'] = df['payment_method'].fillna(mode_payment)
print(f"payment_method missing: {df['payment_method'].isnull().sum()}")
print(f"Filled with: {mode_payment}")
```

##### Fill age group with Unknown
**Reasoning:** The age_group column contained 225 missing values 
representing users who did not complete their profile 
during signup. Since age group cannot be inferred from 
other columns, these were filled with 'Unknown' to 
preserve all records while clearly identifying 
incomplete profiles. This affects only 1.5% of the 
dataset and has minimal impact on demographic analysis.

```python
df['age_group'] = df['age_group'].fillna('Unknown')

print(f"age_group missing: {df['age_group'].isnull().sum()}")
```

#### Fill city with Unknown
**Reasoning:**  The city column contained 300 missing values where 
users did not share their location. Since city cannot 
be reliably inferred from other available columns, 
these were filled with 'Unknown'. This represents 2% 
of the dataset and has minimal impact on geographic 
analysis.

```python
df['city'] = df['city'].fillna('Unknown')

print(f"city missing: {df['city'].isnull().sum()}")
```

#### Fill cash back amount with 0
**Reasoning:** The cashback_amount column contained 668 missing values 
corresponding to Failed and Reversed transactions. Since 
cashback is only awarded on successful transactions under 
Revolut's subscription model, these missing values were 
filled with 0. This is a business logic decision rather 
than a statistical assumption, failed transactions 
genuinely generate no cashback regardless of subscription 
plan.

```python
df['cashback_amount'] = df['cashback_amount'].fillna(0)

print(f"cashback_amount missing: {df['cashback_amount'].isnull().sum()}")
```

#### Outlier detection using IQR method (Percentile)

```python
numerical_cols = ['amount', 'transaction_fee', 'revenue_generated']

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    
    print(f"{col}:")
    print(f"  Q1: {Q1:.2f} | Q3: {Q3:.2f} | IQR: {IQR:.2f}")
    print(f"  Lower bound: {lower:.2f} | Upper bound: {upper:.2f}")
    print(f"  Outliers: {len(outliers):,} ({len(outliers)/len(df)*100:.1f}%)")
    print()
```

#### Decision: Retain Outliers
Outliers identified across amount, transaction_fee and 
revenue_generated represent legitimate high value 
transactions by Premium and Metal plan subscribers. 
Removing them would result in loss
 of valuable business insight. All outliers are retained.
